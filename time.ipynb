# 这是一个关于时间的测试文档

```python
#!/usr/bin/env python
# -*- coding:UTF-8 -*-

'''---------------------------------------
@ Python:3.7
@ OS:Windows10
@ IDE:PyCharm 2019.2
@ Version: 0.1
@ Author: glimmer
---------------------------------------'''
import time
import datetime

httplink = {
    "url": "http://192.168.1.61:8010/radio/getLog",
    "token": "58b820f99e903d3a4844608daba6834a",
    "headers": {
        "User-Agent": "Python-urllib/3.5"
    },
    "value": {
        "limit": "100",
        "start": "1595491921478",
        "end": "1594970769493"
    },
}
ST = datetime.datetime(2020, 8, 1, 19, 21, 10)

now = datetime.datetime.now()
nowtime = now.strftime('%Y-%m-%d %H:%M:%S')
print(nowtime, type(nowtime))
print(now, type(now))
now_stamp = str(time.mktime(now.timetuple()))
print(now_stamp, type(now_stamp))
now_stamp = str(int((time.mktime(now.timetuple()))*1000))
print(now_stamp, type(now_stamp))
print('华丽的分割线'.center(60, '-'))
```

```python
timestamp = time.time()
print(timestamp, type(timestamp))
time_datetime = datetime.datetime.fromtimestamp(timestamp, None)
print(time_datetime, type(time_datetime))
time_datetime = datetime.datetime.fromtimestamp(int(timestamp), None)
print(time_datetime, type(time_datetime))
print('华丽的分割线'.center(60, '-'))
```

```python
print(httplink['value']['start'],httplink['value']['end'])
print(int(httplink['value']['start']),int(httplink['value']['end']))
temp_start = datetime.datetime.fromtimestamp(int(httplink['value']['start'])/1000).strftime('%Y-%m-%d %H:%M:%S')
temp_end = datetime.datetime.fromtimestamp(int(httplink['value']['end'])/1000).strftime('%Y-%m-%d %H:%M:%S')
temp_now = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
print(temp_start, temp_end, temp_now, type(temp_start), type(temp_end), type(temp_now))
print('华丽的分割线'.center(60, '-'))
print(f"ST:{ST}, ST数据类型为：{type(ST)}")
d2 = datetime.datetime.fromtimestamp(int(time.time()), None)
print(f"D2:{d2}, D2类型：{type(d2)}")
delat = d2-ST
print(f"d2-ST:{delat.days}, d2-ST类型：{type(delat.days)}")
```
