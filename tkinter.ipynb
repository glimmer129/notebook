```python
import tkinter as tk
import tkinter.messagebox
import numpy as num
import matplotlib as mat
from time import sleep

global loop
loop = 3
temp = ('top','left','bottom','right')

def main():
    # ------------构架窗口------------
    root = tk.Tk()
    root.title("TEST")
    root.geometry('800x800')
    # ------------构架组件------------
    canva = tk.Canvas(root, bg='green', height=500, width=500)
    image_file = tk.PhotoImage(file='G:\\PYTHON\\python2\\easyGUI\\fruits.gif')
    canva.create_image(250, 250, anchor='center', image=image_file)

    global loop
     #------------GUI输出------------
    if loop < 3:
        loop += 1
    else:
        loop = 0

    canva.pack(side=temp[loop])
    canva.update()
    sleep(1)
    canva.delete()

    #--------------------------------
    root.mainloop()

if __name__=='__main__':
    main()

```
